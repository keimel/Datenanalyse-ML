## Aufgabe 1a) Erstellen sie ein DataFrame

```python
import pandas as pd

city = ["Wien", "Berlin", "Zürich"]
category_names = ["country", "area", "population"]
country = pd.Series(["Österreich", "Deutschland", "Schweiz"], index=city)
area = pd.Series([414.6, 891.85, 87.88], index=city)
population = pd.Series([1805681, 3562166, 378884], index=city)

df1 = pd.concat([country, area, population], axis=1)
df1.columns = category_names

print(df1)
```

Das Grundgerüst des Dataframes ist aufgebaut und mit `stack()` lassen sich die Spaltenindizes "kippen". So erhält man die geforderte Form:

```python
df2 = df1.stack()
print(df2)
```

## 1b) Sortieren Sie nach dem Index

```python
# sortiere nach Flaeche
df3 = df1.sort_values(by="area", ascending=False)
# wechsle die Spaltenordnung
df3 = df3[["area", "country", "population"]]
print(df3.stack())
```

## 1c) Vertauschen von Indizes

```python
print(df3.T.stack())
```

Da in b) bereits nach Fläche sortiert wurde, können  Zeilen und Spalten einfach vertauscht und gestackt werden.

## Aufgabe 2a) Erzeugen Sie einen DataFrame mit einem Index (nicht nachträglich setzen, Index direkt erzeugen), der aus Vornamen besteht, sowie eine Spalte für das Gewicht und einer Spalte für die Körpergröße.

```python
import pandas as pd

names = ["Siegfried", "Henry", "Sarah", "Elke", "Lulu", "Vera", "Toni", "Maria", "Chris"]
weight = [403, 65, 58, 58, 45, 43, 99, 68, 60]
hight = [180, 179, 165, 172, 154, 150, 189, 176, 175]

data = {"name": names, "Gewicht":weight, "Größe":hight}
dfA = pd.DataFrame(data, columns=["Gewicht", "Größe"], index=names)
print(dfA)
```

## 2b) Extrahieren Sie alle Zeilen, deren BMI im normalen Bereich liegt, d.h. zwischen 18.5 & 25.
### Es gilt: BMI = Gewicht (kg)/ (Größe in m)².

```python
print(dfA.query("Gewicht / ((Größe/100) ** 2) > 18.5 and Gewicht / ((Größe/100) ** 2) < 25"))
```

## 2c) Geben Sie alle Zeilen aus, in deren Namen ein kleines "e" vorkommt.

```python
print(dfA[dfA.index.str.contains('e')])
```

## 2d) Fügen Sie eine Spalte mit dem BMI an.

```python
dfA["BMI"] = dfA["Gewicht"] / ((dfA["Größe"] / 100) ** 2)
print(dfA)
```

## 2e) Sortieren Sie nach dem BMI

```python
print(dfA.sort_values(by="BMI", ascending=False))
```

## 2f) Geben Sie nun alle Zeilen aus, deren BMI-Werte unter 20 liegen und deren Name ein kleines "e" enthält

```python
cond1 = dfA.BMI < 20
cond2 = dfA.index.str.contains('e')
print(dfA[cond1 & cond2])
```
