a) Erzeugen Sie eine neue Datei mit den Spaltennamen „land“, “area“, „female“, „male“, „ population“ und „density“. Die density ist die Anzahl der Einwohner pro Quadratkilometer. 

```python
file2 = "bundeslaender.txt"
columns = ['land', 'area', 'female', 'male']
states = pd.read_csv(file2, #demoStates
                     sep = " ",
                     quotechar = "'",
                     header = 0) # erste Zeile überspringen
states = states[columns] # Spaltenordnung "female - male" anpassen
states['population'] = states.female + states.male
states['density'] = round((states.population / states.area) * 1000)

print(states)
```

b) Geben Sie alle Zeilen aus mit einer größeren weiblichen Bevölkerungszahl als die der Männer.
Wie viele sind es?

```python
print(states.query("female > male"))
```

c) Geben Sie alle Zeilen aus, deren Bevölkerung größer als 1000 Einwohner pro Quadratkilometer ist.

```python
print(states.query("density >= 1000"))
```
