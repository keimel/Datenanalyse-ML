# Aufgabe 1
Laden Sie den iris-Datensatz und zeichnen Sie vier Histogramme jeweils der vier numerischen Merkmale für die
drei Objektklassen setosa, versicolor, und virginica.  
Färben Sie nach den Objektklassen ein.  
Erzeugen Sie die folgenden Ausgaben:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('iris.csv', sep=',') # Daten einlesen

# Aufspalten in die verschidenen Sorten
## boolean DF -> Sorten DF
setosa_df     = df[df["variety"] == "Setosa"]
versicolor_df = df[df["variety"] == "Versicolor"]
virginica_df  = df[df["variety"]== "Virginica"]

# Sepal Länge der einzelnen Sorten sammeln
sepal_length_setosa     = setosa_df["sepal.length"]
sepal_length_versicolor = versicolor_df["sepal.length"]
sepal_length_virginica  = virginica_df["sepal.length"]

# Boxplots der Sepal Längen zeichnen
plt.hist(sepal_length_setosa, color="blue", label="Setosa")
plt.hist(sepal_length_versicolor, color="red", label="Versicolor")
plt.hist(sepal_length_virginica, color="green", label="Virginica")

plt.xlabel("sepal length (cm)")

plt.legend()
plt.show()
```

```python
# Blüttenbreite der Sorten
petal_with_setosa     = setosa_df["petal.width"]
petal_with_versicolor = versicolor_df["petal.width"]
petal_with_virginica  = virginica_df["petal.width"]

plt.hist(petal_with_setosa, color="blue", label="Setosa")
plt.hist(petal_with_versicolor, color="red", label="Versicolor")
plt.hist(petal_with_virginica, color="green", label="Virginica")

plt.xlabel("petal width (cm)")

plt.legend()
plt.show()
```

```python
petal_length_setosa = setosa_df["petal.length"]
petal_length_versicolor = versicolor_df["petal.length"]
petal_length_virginica = virginica_df["petal.length"]

plt.hist(petal_length_setosa, color="blue", label="Setosa")
plt.hist(petal_length_versicolor, color="red", label="Versicolor")
plt.hist(petal_length_virginica, color="green", label="Virginica")

plt.xlabel("petal length (cm)")

plt.legend()
plt.show()
```

```python
sepal_wdth_setosa = setosa_df["sepal.width"]
sepal_wdth_versicolor = versicolor_df["sepal.width"]
sepal_wdth_virginica = virginica_df["sepal.width"]

plt.hist(sepal_wdth_setosa, color="blue", label="Setosa")
plt.hist(sepal_wdth_versicolor, color="red", label="Versicolor")
plt.hist(sepal_wdth_virginica, color="green", label="Virginica")

plt.xlabel("sepal width (cm)")

plt.legend()
plt.show()
```

- Berechnen Sie den Korrelationskoeffizient für die beiden Merkmalspaare aus Aufgabe 1 (Kelchblattlänge u. Kelchblattbreite,
Blütenblattlänge und Blütenblattbreite. Erzeugen Sie die Ausgaben:
